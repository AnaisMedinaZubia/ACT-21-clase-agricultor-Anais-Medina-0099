```python
class Agricultor:
    def __init__(self, nombre, telefono, edad):
        # Estos son los 3 atributos (datos)
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Preparar el suelo
    def preparar_suelo(self):
        print(f"{self.nombre} está preparando la tierra con el arado.")

    # Función 2: Sembrar
    def sembrar(self):
        print(f"{self.nombre} está colocando las semillas en los surcos.")

    # Función 3: Regar
    def regar(self):
        print(f"{self.nombre} está activando el sistema de riego para sus cultivos.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1 con sus datos de ejemplo
agricultor1 = Agricultor("Juan Pérez", "555-0123", 45)

# Objeto 2 con sus datos de ejemplo
agricultor2 = Agricultor("María García", "555-0987", 38)

# --- Ejecución de las funciones y mensajes ---

print(f"--- Ficha de {agricultor1.nombre} ---")
print(f"Teléfono: {agricultor1.telefono}, Edad: {agricultor1.edad}")
agricultor1.preparar_suelo()
agricultor1.sembrar()
agricultor1.regar()

print("\n" + "-"*30 + "\n")

print(f"--- Ficha de {agricultor2.nombre} ---")
print(f"Teléfono: {agricultor2.telefono}, Edad: {agricultor2.edad}")
agricultor2.preparar_suelo()
agricultor2.sembrar()
agricultor2.regar()

print("Programa terminado por Anais Medina NL = 36")
```
